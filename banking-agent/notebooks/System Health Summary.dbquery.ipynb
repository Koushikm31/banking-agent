```sql
SELECT 
    'Request Count' as metric_name,
    COUNT(*) as value,
    'requests' as unit
FROM banking_agent_db.request_logs
WHERE timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR

UNION ALL

SELECT 
    'Avg Latency (ms)',
    ROUND(AVG(latency_ms), 0),
    'ms'
FROM banking_agent_db.request_logs
WHERE timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR

UNION ALL

SELECT 
    'Success Rate (%)',
    ROUND(100.0 * SUM(CASE WHEN status = 'success' THEN 1 ELSE 0 END) / COUNT(*), 1),
    '%'
FROM banking_agent_db.request_logs
WHERE timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR

UNION ALL

SELECT 
    'Model Accuracy (%)',
    ROUND(100.0 * AVG(metric_value), 1),
    '%'
FROM banking_agent_db.metrics
WHERE metric_name = 'accuracy' 
  AND timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR

UNION ALL

SELECT 
    'Total Cost ($)',
    ROUND(SUM(cost), 2),
    '$'
FROM banking_agent_db.cost_tracking
WHERE timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR
```
