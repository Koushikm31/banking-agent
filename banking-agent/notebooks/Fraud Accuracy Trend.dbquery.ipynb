```sql
SELECT 
    date_format(timestamp, 'HH:00') as hour,
    dimension_1 as model_name,
    ROUND(100.0 * AVG(metric_value), 2) as accuracy_percent,
    COUNT(*) as predictions,
    SUM(CASE WHEN metric_value = 1 THEN 1 ELSE 0 END) as correct_predictions
FROM banking_agent_db.metrics
WHERE metric_name = 'accuracy' 
  AND timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR
GROUP BY date_format(timestamp, 'HH:00'), dimension_1
ORDER BY hour DESC, accuracy_percent DESC
```
