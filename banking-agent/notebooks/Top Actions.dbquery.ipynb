```sql
SELECT 
    action,
    COUNT(*) as usage_count,
    ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM banking_agent_db.request_logs 
                               WHERE timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR), 2) as percent_of_total
FROM banking_agent_db.request_logs
WHERE timestamp >= CURRENT_TIMESTAMP - INTERVAL 24 HOUR
GROUP BY action
ORDER BY usage_count DESC
LIMIT 10
```
